```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
exe = sns.load_dataset("titanic")
```

```python
print(exe.head())
```

```python
sns.set_style("white")
sns.scatterplot(data=exe,x="age",y="embark_town",hue="class",style="sex",palette="Set1")
sns.set_context("notebook")
sns.despine(bottom=False)
```

```python
sns.stripplot(data=exe,x="age",y="embark_town",hue="sex",palette="Set1",dodge=True)
```

```python
sns.swarmplot(data=exe,x="age",y="embark_town",hue="sex")
sns.despine(left=True,bottom=True)
```

```python
sns.histplot(data=exe,x="class",hue="sex",multiple="stack")
```

```python
sns.regplot(data=exe,x="fare",y="age",scatter=False)
```

```python
sns.lineplot(data=exe,x="age",y="fare",hue="sex",style="sex")
```

```python
sns.jointplot(data=exe,x="age",y="fare",kind="hist")
```

```python
sns.barplot(data=exe,x="who",y="age",hue="alone",palette=["lightblue","red"])
```

```python
sns.countplot(data=exe,x="who",hue="sex",palette="Set2")
```

```python
sns.boxplot(data=exe,x="who",y="age",hue="alone",palette=["red","blue"])
```

```python
sns.violinplot(data=exe,x="who",y="age",hue="alone",split=True)
sns.swarmplot(data=exe,x="who",y="age",palette=["pink"])
```

```python
sns.kdeplot(data=exe,x="age",hue="sibsp",palette=["red","blue"])
```

```python
exe.head()
```

```python
columns=["age","sibsp","parch","fare"]
sns.heatmap(data=exe[columns].corr(),annot=True,cmap="Blues",vmin=-0.2,vmax=1.2,linewidths=0.01,linecolor="Red")
```

```python
sns.rugplot(data=exe,y="fare",color=["Green"],height=1)
```

```python
sns.pairplot(data=exe[columns],hue="age",palette="Set2",diag_kind="hist")
```
